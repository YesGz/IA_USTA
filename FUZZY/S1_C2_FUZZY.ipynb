# Importando Librerias

```python
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import numpy as np
import matplotlib.pyplot as plt
```

# Ejercicio 1 Graficar Variable Temperatura

```python
Temp=ctrl.Antecedent(np.arange(80,110,1),"Temperatura")
Temp['Alto']=fuzz.trapmf(Temp.universe,[80,80,90,100])
Temp['Muy Alto']=fuzz.trapmf(Temp.universe,[90,100,110,110])
Temp.view()
```

# Ejercicio 2 Tres Funciones de pertenecia

```python
Temp2=ctrl.Antecedent(np.arange(60,110,1),"Temperatura")
Temp2['Normal']=fuzz.trapmf(Temp2.universe,[60,60,70,80])
Temp2['Alto']=fuzz.trapmf(Temp2.universe,[70,80,90,100])
Temp2['Muy Alto']=fuzz.trapmf(Temp2.universe,[90,100,110,110])
Temp2.view()
```
